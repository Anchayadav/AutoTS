```python
import pandas as pd
import yfinance  as yf
from datetime import datetime

start_date=datetime.now() - pd.DateOffset(month=3)
end_date=datetime.now()
tikers= ['AAPL','MSFT','NFLX','GOOG']
df_list =[]
for tiker in tikers:
  data=yf.download(tiker,start=start_date,end=end_date)
  df_list.append(data)



df=pd.concat(df_list,keys=tikers,names=['Tiker','Date'])
print(df.head())
```

```python
df=df.reset_index()
print(df.head())
```

```python
import plotly.express as px
fig=px.line(df,x='Date',y='Close',
            color='Tiker',
            title="stock market performance for the last 3 month")
fig.show()
```

```python
fig=px.area(df,x='Date',y='Close',color='Tiker',
            facet_col='Tiker',
            labels={'Date':'Date','Close':'Closeing Price','Tiker':'Company'},
            title='stock prices for apple ,Microsoft,netflix,google')
fig.show()
```

```python
df['MA10']= df.groupby('Tiker')['Close'].rolling(window=10).mean().reset_index(0,drop=True)
df['MA20']= df.groupby('Tiker')['Close'].rolling(window=20).mean().reset_index(0,drop=True)
for tiker,group in df.groupby('Tiker'):
  print(f'Moving Average for {tiker}')
  print(group[['MA10','MA20']])
```

```python
for tiker,group in df.groupby('Tiker'):
  fig=px.line(group,x='Date',y=['Close','MA10','MA20'],
  title=f"{tiker} Moving Averges")
fig.show()
```

```python
df['Volatility'] =df.groupby('Tiker')['Close'].pct_change().rolling(window=10).std().reset_index(0,drop=True)
fig=px.line(df,x='Date',y='Volatility',
            color='Tiker',
            title='Volatility of all companies')
fig.show()
```

```python
apple =df.loc[df['Tiker']=='AAPL',['Date','Close']].rename(columns={'Close':'AAPL'})
microsoft=df.loc[df['Tiker']=='MSFT',['Date','Close']].rename(columns={'Close':'MSFT'})
df_corr=pd.merge(apple,microsoft,on='Date')
fig=px.scatter(df_corr, x='AAPL',y='MSFT',
               trendline='ols',
               title='Correlation between apple an microsoft')
fig.show()
```
