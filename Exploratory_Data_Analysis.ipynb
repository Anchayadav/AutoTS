```python
fig=go.Figure()

fig.add_trace(go.Scatter(x=data.index,y=data['Likes'],name='like'))
fig.add_trace(go.Scatter(x=data.index,y=data['Saves'],name='Saves'))
fig.add_trace(go.Scatter(x=data.index,y=data['Follows'],name='Follow'))

fig.update_layout(title='Metrics over time',
                  xaxis_title='Date',
                  yaxis_title='Count')
fig.show()
```

```python
reach_sources = ['From Home','From Hashtags','From Explore','From Other']
reach_counts = [data[source].sum() for source in reach_sources]

colors = ['#FFB6C1','#87CEFA','#90EE90','#FFDAB9']
fig=px.pie(data_frame =data ,
           names= reach_sources,
           values=reach_counts,
           title="reach from ythe differnet sources",
           color_discrete_sequence=colors,
           )

fig.show()
```

```python
engagement_metrics = ['Saves','Comments','Shares','Likes']
engement_count = [data[metrics].sum() for metrics in engagement_metrics]

colors = ['#FFB6C1','#87CEFA','#90EE90','#FFDAB9']
fig=px.pie(data,names=engagement_metrics ,values=engement_count,
           title='engement fro the other sources',
           color_discrete_sequence=colors)
fig.show()
```

```python
correlation_matrix=data.corr()
fig = go.Figure(data=go.Heatmap(z=correlation_matrix.values,
                                x=correlation_matrix.columns,
                                y=correlation_matrix.index,
                                zmin=-1,
                                zmax=1,
                                colorscale='RdBu'))
fig.update_layout(title="correlation matrix",
                  xaxis_title='features',
                  yaxis_title='features')
fig.show()
```

```python
all_hashtags=[]

for row in data['Hashtags']:
  hashtags=str(row).split()
  hashtags = [tag.strip for tag in hashtags]
  all_hashtags.extend(hashtags)

hastag_distribution = pd.Series(all_hashtags).value_counts().reset_index()
hastag_distribution.columns=['Hashtags','Count']

fig_hastags = px.bar(hastag_distribution,x='Hashtags',
           y='Count',title='Distribution of Hashtgas')
fig_hastags.show()
```

```python
hashtags_like = {}
hashtag_impressions={}

for index, row in data.iterrows():
  hashtags=str(row['Hashtags']).split()
  for hashtags in hashtags:
    hashtags = hashtags.strip()
    if hashtags not in hashtags_like:
      hashtags_like[hashtags] =0
      hashtag_impressions[hashtags] = 0
      hashtags_like[hashtags]  +=row['Likes']
      hashtag_impressions[hashtags]  +=row['Impressions']

likes_distribution=pd.DataFrame(list(hashtags_like.items()),columns=['Hashtags','Likes'])
impression_distribution=pd.DataFrame(list(hashtag_impressions.items()),columns=['Hashtags','Impressions'])

fig_likes=px.bar(likes_distribution,x='Hashtags',y='Likes',
                 title='Likes Diostribution of each hashtsg')

fig_impressions=px.bar(impression_distribution,x='Hashtags',y='Impressions',
                 title='Impressions Diostribution of each hashtsg')
fig_likes.show()
fig_impressions.show()
```

```python
all_hastgag =[]

for row in data['Caption'] :
  hastags = str(row).split()
  hastasg = [tag.strip() for tag in hastags]
  all_hastgag.extend(hastasg)

caption_dixtribution = pd.Series(all_hastgag).value_counts().reset_index()
caption_dixtribution.columns =['Caption','Count']

fig_caption = px.bar(caption_dixtribution,x='Caption',
                     y='Count',title='caption for distribution for ')
fig_caption.show()
```
