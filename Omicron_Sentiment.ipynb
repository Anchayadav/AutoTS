```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import  SentimentIntensityAnalyzer
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator\

data=pd.read_csv('omicron.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
```

```python
import nltk
import string
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
stemmer=nltk.SnowballStemmer('english')
stopword=set(stopwords.words('english'))

def clean(text):
  text = str(text).lower()
  text =re.sub('\[ .*?\]','',text)
  text=re.sub('https?://\S+|www\.S+','',text)
  text=re.sub('\n','',text)
  text =re.sub('[%]' %re.escape(string.punctuation),'',text)
  text=re.sub('<.?*> +','',text)
  text=[word for word in text.split('')if word not in stopword]
  text=' '.join(text)
  text =[stemmer.stem(word)for word in text.split()]
  text=' '.join(text)
  return text

  data['text'] =data['text'].apply(clean)
```

```python
text = ' '.join(i for i in data.text)
stopwords=set(STOPWORDS)
wordclod=WordCloud(stopwords=stopwords,background_color='white').generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordclod,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
text =' '.join(i for i in data.hashtags)
stopwords =set(STOPWORDS)
wordcloud =WordCloud(stopwords=stopwords,background_color='white').generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
nltk.download('vader_lexicon')
sentiment =SentimentIntensityAnalyzer()
data['Positive'] =[sentiment.polarity_scores(i)['pos']for i in data['text']]
data['Negative']=[sentiment.polarity_scores(i)['neg']for i in data['text']]
data['Neutral'] =[sentiment.polarity_scores(i)['neu']for i in data['text']]
data = data[['text','Positive','Negative','Neutral',]]
print(data.head())
```

```python
x=sum(data['Positive'])
y=sum(data['Negative'])
z=sum(data['Neutral'])

def  sentiment_analysis(a,b,c):
  if (a>b) and (a>c):
    print("Positive")
  elif (b>a) and (b>c):
    print("Negative")

  else:
    print("Neutral")
sentiment_analysis(x,y,z)
```
