```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from nltk.corpus import stopwords

data=pd.read_csv('Myntra kurtis.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data=data.drop('Image',axis=1)
```

```python
data =data.dropna()
```

```python
print(data.shape)
```

```python

text = ' '.join( i for i in data['Brand Name'])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords =stopwords,background_color = 'white').generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud,interpolation ='bilinear')
plt.axis('off')
plt.show()
```

```python
highest_rate = data.sort_values(by=['Product Ratings'],
                           ascending=False)
highest_rate =highest_rate.head(10)
print(highest_rate[['Product Info','Product Ratings','Brand Name']])
```

```python
mr=data['Product Ratings'].mean()
m=data['Number of ratings'].quantile(0.9)
n=data['Number of ratings']
a=data['Product Ratings']
data['Score'] =(n/(n+m)*a) +(m/(m+n)*mr)

recommendation= data.sort_values('Score',ascending=False)

print(recommendation[['Brand Name','Product Info','Product Ratings','Score','Selling Price','Price']].head(10))
```
