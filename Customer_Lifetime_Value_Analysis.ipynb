```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.templates.default = 'plotly_white'
import plotly.graph_objects as go
data=pd.read_csv('customer_acquisition_data.csv')
print(data.head())
```

```python
fig=px.histogram(data,x='cost',
                 nbins=20,
                 title='ditribution of the acquition  cost')
fig.show()
```

```python
ig=px.histogram(data,x='revenue',
                 nbins=20,
                 title='ditribution of revenue')
fig.show()
```

```python
cost_by_channel = data.groupby('channel')['cost'].mean().reset_index()
fig=px.bar(cost_by_channel,
           x='channel',
           y='cost',
           title='channel by the cost'
           )
fig.show()
```

```python
print(data.isnull().sum())
```

```python
conversion_by_channel = data.groupby('channel')['conversion_rate'].mean().reset_index()
fig=px.bar(conversion_by_channel,x='channel',
           y='conversion_rate',
           title='converion rate by channel')
fig.show()
```

```python
revenue_by_channel = data.groupby('channel')['revenue'].sum().reset_index()
fig=px.pie(revenue_by_channel,values='revenue',
           names='channel',
           hole=0.25,
           title='Total revenue by the channel',
           color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

```python
data['roi'] = data['revenue'] / data['cost']
roi_by_channel = data.groupby('channel')['roi'].mean().reset_index()
fig=px.bar(roi_by_channel,x='channel',
           y='roi',
           title=' return on investment(ROI) by the channel')
fig.show()
```

```python
data['cltv'] = (data['revenue'] -  data['cost']) * data['conversion_rate'] / data['cost']
channel_cltv = data.groupby('channel')['cltv'].mean().reset_index()
fig=px.bar(channel_cltv,
           x='channel',
           y='cltv',
           color='channel',
           title='costomer life time value  by channel')
fig.update_xaxes(title='channel')
fig.update_yaxes(title='cltv')
fig.show()
```

```python
subset = data.loc[data['channel'].isin(['social media','referral'])]
fig=px.box(subset , x='channel',
           y='cltv',
           title='cltv distribution by channel')
fig.update_xaxes(title='Channel')
fig.update_yaxes(title='referral')
fig.update_layout(legend_title='Channel')
fig.show()
```
