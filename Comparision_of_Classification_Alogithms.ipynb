```python
import numpy as np
import pandas as pd
import numpy
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import classification_report

data=pd.read_csv('social.csv')
print(data.head())
```

```python
x=np.array(data[['Age','EstimatedSalary']])
y=np.array(data[['Purchased']])

xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.2,random_state=42)

decisiontree= DecisionTreeClassifier()
logistice = LogisticRegression()
Knearsest= KNeighborsClassifier()
#svm_classier = SVC()
bernouli_naiyebayes =BernoulliNB()
passiveAggresive= PassiveAggressiveClassifier()

Knearsest.fit(xtrain,ytrain)
decisiontree.fit(xtrain,ytrain)
logistice.fit(xtrain,ytrain)
passiveAggresive.fit(xtrain,ytrain)

data1={'Classification Algoritms':['Knn Classifier','Decision Tree',
                                   'Logistic Regression','Passive Aggressive'],
       "score":[Knearsest.score(x,y),decisiontree.score(x,y),
                logistice.score(x,y),passiveAggresive.score(x,y)] }
score = pd.DataFrame(data1)
score
```
